```python
df = pd.read_csv("H-MERIDIANHEALTH_cleaned.csv")
df["readmitted"] = (df["readmission_30day"] == "Yes").astype(int)
```

# Meridian Health EDA

```python
sns.barplot(data=df, y="primary_diagnosis", x="readmitted", color='green')
plt.title("Readmission Rate by Diagnosis")
plt.savefig('chart1.png')
plt.show()
```

```python
sns.barplot(data=df, y="discharge_disposition", x="readmitted")
plt.title("Readmission Rate by Discharge Disposition")
plt.savefig("chart2.png")
plt.show()
```

```python
sns.barplot(data=df, y="admission_type", x="readmitted", color='red')
plt.title("Readmission Rate by Admission Type")
plt.savefig("chart3.png")
plt.show()
```

```python
sns.barplot(data=df, y="follow_up_scheduled", x="readmitted", color='orange')
plt.title("Readmission Rate by Follow-Up Scheduled")
plt.savefig("chart4.png")
plt.show()
```

```python
sns.barplot(data=df, y="medication_reconciliation_completed", x="readmitted", color='purple')
plt.title("Readmission Rate by Medication Reconciliation")
plt.savefig("chart5.png")
plt.show()
```

```python
sns.barplot(data=df, y="post_discharge_call_made", x="readmitted", color='pink')
plt.title("Readmission Rate by Post-Discharge Call")
plt.savefig("chart6.png")
plt.show()
```

```python
sns.barplot(data=df, y="insurance_type", x="readmitted", color='gray')
plt.title("Readmission Rate by Insurance Type")
plt.savefig("chart7.png")
plt.show()
```

```python
sns.barplot(data=df, y="hospital_location_type", x="readmitted", color='yellow')
plt.title("Readmission Rate by Hospital Location")
plt.savefig("chart8.png")
plt.show()
```

```python
sns.histplot(data=df, x="patient_age", hue="readmission_30day", kde=True)
plt.title("Age Distribution")
plt.savefig("chart9.png")
plt.show()
```

```python
sns.histplot(data=df, x="length_of_stay_days", hue="readmission_30day")
plt.title("Length of Stay Distribution")
plt.savefig("chart10.png")
plt.show()
```

```python
sns.histplot(data=df, x="comorbidity_count", hue="readmission_30day", bins=15)
plt.title("Comorbidity Count Distribution")
plt.savefig("chart11.png")
plt.show()
```

```python
sns.histplot(data=df, x="ed_visits_past_year", hue="readmission_30day", bins=15)
plt.title("ED Visits Past Year Distribution")
plt.savefig("chart12.png")
plt.show()
```

```python
sns.scatterplot(data=df, x="ed_visits_past_year", y="social_determinants_risk_score", hue="readmission_30day")
plt.title("ED Visits vs Social Risk Score")
plt.savefig("chart13.png")
plt.show()
```

```python
cols = ["comorbidity_count", "patient_age", "length_of_stay_days",
        "social_determinants_risk_score", "ed_visits_past_year", "readmitted"]
sns.heatmap(df[cols].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.savefig("chart14.png")
plt.show()
```
